```python
if __name__ == '__main__':
    keyword = ['river','lake']
   # keyword =['lake','reservoir','dam','river','wetlands','lake'] # marsh


    flicker = flickrapi.FlickrAPI (api_key,api_secret, format = 'parsed-json')
 
   # Box= [minimum_longitude,minimum_latitude,maximum_longitude,maximum_latitude= -81.2387925042645662,42.4830943145320887,
            #-77.3770526963301108,45.2951482390430513]

 # keeping has_geo parameter as 1, accuracy =16, min_date_taken= "2020/12/01" and max_date_taken = "2020/12/31"

    response = flicker.photos.search (
        text = keyword,
        media = 'photos',
        sort = 'relevance',
        safe_search = 1,
        bbox="-81.2387925042645662,42.4830943145320887,-77.3770526963301108,45.2951482390430513",
        has_geo = 1,
        tag_mode='any',
        tags=keyword,
        accuracy=16,
        per_page=500,
        geo_context=2,
        lat="-81.23879250426457",
        lon="45.29514823904305",
        extras='url_c,date_upload, date_taken, owner_name,geo,tags,views',
        min_taken_date="2020-01-01",
        max_taken_date="2020-12-31",
        
        )
    photos = response ['photos']
   
    try:
        out_dirpath = os.path.join ('image_data', keyword)
        os.makedirs (out_dirpath, exist_ok = True)
        for photo in photos ['photo']:
            url = photo ['url_c']
            out_path = os.path.join (out_dirpath, photo ['id'] + '.jpg')
           # print ('getting image ...', url)
            urlretrieve (url, out_path)
            time.sleep (1)
    except Exception as e:
        traceback.print_exc ()
    with open ('p{}1param.json'.format (keyword),'w') as f:
        json.dump (photos, f, sort_keys = True, indent = 4)
        
for i        
print("done") 
```

```python
# converting json file to csv and excel
# read json

with open(r'flickrdata.json', encoding='utf-8') as f:
    data = json.loads(f.read())
#data=pd.read_csv('priver1.csv')

fname="out_river1.csv"

with open(fname,"w",encoding='utf-8') as file:
    csv_file=csv.writer(file)
    csv_file.writerow(["accuracy","date_taken","dateupload","id","views","latitude","longitude","owner","owner_name","placeid","tags","title","url","woeid"])
    for item in data["photo"]:
        csv_file.writerow([item["accuracy"],item['datetaken'],item['dateupload'],item['id'],item['views'],
                          item['latitude'],item['longitude'],item['owner'],item['ownername'],
                          item['place_id'],item['tags'],item['title'],item['url_c'],item['woeid']])
```
